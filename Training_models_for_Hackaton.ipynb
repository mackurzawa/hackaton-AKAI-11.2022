## Obtaining data

```python
import tensorflow as tf
```

```python
train_data= tf.keras.preprocessing.image_dataset_from_directory('splittingHairColors/Train',
                                                                               label_mode='binary',
                                                                               image_size=(224, 224))

test_data= tf.keras.preprocessing.image_dataset_from_directory('splittingHairColors/Test',
                                                                              label_mode='binary',
                                                                              image_size=(224, 224),
                                                                              shuffle=False)
```

```python
base_model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, 3, activation='relu', input_shape=(224, 224, 3)),
    tf.keras.layers.MaxPool2D(),
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPool2D(),
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPool2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
base_model.summary()
```

```python
base_model.compile(loss='binary_crossentropy',
                   optimizer='adam',
                   metrics='accuracy')
```

```python
epochs = 2
# checkpoint_path = 'checkpoint/feature_extraction'
# checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                        #  save_weights_only=True,
                                                        #  monitor='accuracy',
                                                        #  save_best_only=True)
# early_stop_callback = tf.keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)
```

```python
history = base_model.fit(train_data,
                        steps_per_epoch=len(train_data),
                        epochs=epochs,
                        validation_data=test_data,
                        validation_steps=len(test_data),
                        # callbacks=[checkpoint_callback,
                        #             early_stop_callback]
                         )
```
